<h1>Read DataSet<h1>

```python
import pandas as pd

data = pd.read_csv ('DataSet.csv')   
df = pd.DataFrame(data, columns= ['Tweet'])
print (df)
```

```python
# mylist = df['Tweet'].tolist()
mylist = df['Tweet']
re_mylist=[]
ru_mylist=[]
rd_mylist=[]
rp_mylist=[]
rew_mylist=[]
my_new_list=[]
mylist[:10]
```

<h1>Remove email<h1>

```python
mylist=mylist.str.replace(r'\S*@\S*\s?|^.+@[^\.].*\.[a-z]{2,}$','')
mylist
```

<h1>Remove URL<h1>

```python
mylist=mylist.str.replace(r'https?:\/\/.*[^\w\r\n\s\d+]*','')
mylist
```

<h1>Remove numbers<h1>

```python
mylist=mylist.str.replace(r'^\d+\s|\d+(\.\d+)?|\s\d+\s|\s\d+$','')
mylist
```

<h1>Remove Punc<h1>

```python
mylist=mylist.str.replace(r'[^\w\s]','')
mylist
```

<h1>Remove English Words<h1>

```python
mylist=mylist.str.replace(r'\s*[A-Za-z]+\b|^[^A-Za-z]+?','')
mylist
```

<h1>Word Tokenize<h1>

```python
from nltk.tokenize import word_tokenize
for i in mylist:
    mylist=word_tokenize(i)
    my_new_list.append(mylist)
my_list=my_new_list
print(len(my_list))
my_list
```

<h1>Remove Stopwords<h1>

```python
from nltk.corpus import stopwords
ar_stop_words = set(stopwords.words("arabic"))
ar=[]
for i in my_list:
    ar_cleaned_words = [word for word in i if word not in ar_stop_words]
    ar.append(ar_cleaned_words)
print(len(ar))
ar
```

<h1>Remove Frequent Words (Manully) :D<h1>

```python
from nltk.probability import FreqDist
ar_freq_words=["ู","ุบูุบู","ููุณู","ูุธุงู","ุงุฐุง","ุงู","ูุงุฑู","ุจูุชุฑ","ูููุชูู","ูููุงุฑู","ุบูุชุดู","ุง","ุงููู","ุจุฏูู","ูุซู","ูููุฏูุฒ","ููุฏุฑุฑ","ุฑูุฌุฑ","ุชุฑุงูุจ","ูุงู","ุงูู","ูุงุฏุงู","ู","ุงูุงุฑุฌูุชูู","ุงูุชูุณ","ู","ุงูู","ุงูุฌุฏูุฏ","ุน","ูุชุดู","ุดูุก"]
c=0
for i in ar:
    for j in i:
        if j in ar_freq_words:
            i.remove(j)
            c=c+1
print(len(ar))
print(c)
ar
```

```python
last=[]
for i in ar:
    listToStr = ' '.join([str(elem) for elem in i])
    last.append(listToStr)
last
```

```python
from nltk.tokenize import word_tokenize
all_words = []
for message in last:
    words = word_tokenize(message)
    for w in words:
        all_words.append(w)
        
all_words = FreqDist(all_words)
all_words
```
