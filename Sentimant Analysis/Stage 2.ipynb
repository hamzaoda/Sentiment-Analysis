```python
import pandas as pd

data = pd.read_csv ('CleanedDataSet.csv') 
data.drop('Unnamed: 0', inplace=True, axis=1)
data.drop('Unnamed: 0.1', inplace=True, axis=1)
df = pd.DataFrame(data)
print (df)
```

```python
mylist = df['0'].map(str)
print(type(mylist))
mylist
```

<h1>Add Snow_Ball_Stemmer<h1>

```python
from nltk.stem import SnowballStemmer
s_stemmer = SnowballStemmer('arabic')
sb_stemmer = mylist.apply(lambda x: ' '.join(
    s_stemmer.stem(term) for term in x.split()))
sb_stemmer
df['Snow_Ball_Stemmer']=sb_stemmer
df.to_csv('CleanedDataSet.csv')
df
```

<h1>Add ISRI_Stemmer<h1>

```python
import nltk
from nltk.stem.isri import ISRIStemmer
st = ISRIStemmer()
ISRI_stemmer = mylist.apply(lambda x: ' '.join(
    st.stem(term) for term in x.split()))
df['ISRI_Stemmer']=ISRI_stemmer
df.to_csv('CleanedDataSet.csv')
df
```

<h1>Add ISRI_suffix(lemma)<h1>

```python
ISRI_suf = mylist.apply(lambda x: ' '.join(
    st.suf32(term) for term in x.split()))
df['ISRI_suffix(lemma)']=ISRI_suf
df.to_csv('CleanedDataSet.csv')
df
```

<h1>Add ARLSTem_Stems<h1>

```python
from nltk.stem.arlstem import ARLSTem
stems = mylist.apply(lambda x: ' '.join(
    ARLSTem().stem(term) for term in x.split()))
df['ARLSTem_Stems']=stems
df.to_csv('CleanedDataSet.csv')
df
```

<h1>Add ARLSTem_Verbs<h1>

```python
verbs = mylist.apply(lambda x: ' '.join(
    ARLSTem().verb(term) for term in x.split()))
df['ARLSTem_Verbs']=verbs
df.to_csv('CleanedDataSet.csv')
df
```
