```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
import pandas as pd

data = pd.read_csv ('emotion.csv')  
df = pd.DataFrame(data)
x=pd.Series(df.Arabic_Term.values,index=df.Sentiment_Score).to_dict()
x
```

```python
positive_emt=[]
negitive_emt=[]
neutral_emt=[]
for key in x:
    if key > 0.5:
        positive_emt.append(x[key])
    elif key <-0.5:
        negitive_emt.append(x[key])
    else:
        neutral_emt.append(x[key])  
```

```python
def count_neg(text):
    neg = 0
    for w in word_tokenize(text):
        if w in negitive_emt:
            neg +=1

    return neg
```

```python
def count_pos(text):
    pos = 0
    for w in word_tokenize(text):
        if w in positive_emt:
            pos +=1
            
    return pos
```

```python
def count_neutral(text):
    neu = 0
    for w in word_tokenize(text):
        if w in neutral_emt:
            neu +=1
            
    return neu
```

<h1>Snow_Ball_Stemmer<h1>

```python
import pandas as pd

data = pd.read_csv ('CleanedDataSet.csv',usecols=['Snow_Ball_Stemmer','feeling'])  
df = pd.DataFrame(data)
print (df)
```

```python
from nltk import *
df['Snow_Ball_Stemmer'] = df['Snow_Ball_Stemmer'].astype(str)
def pre_process(df):
    df['feeling']
    df['pos'] = df['Snow_Ball_Stemmer'].apply(count_pos)
    df['neg'] = df['Snow_Ball_Stemmer'].apply(count_neg)
    df['neu'] = df['Snow_Ball_Stemmer'].apply(count_neutral)
    return df
pre_process(df)
```

```python
X = df[['pos', 'neg','neu']]
```

```python
from sklearn.preprocessing import LabelEncoder
l = LabelEncoder()
y = l.fit_transform(df[['feeling']])
```

```python
from sklearn import model_selection
X_train , X_test, y_train , y_test = model_selection.train_test_split(X,y, test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
clf.fit(X_train , y_train)
y_predict = clf.predict(X_test)
```

```python
from sklearn.metrics import *

print('accuracy_score', accuracy_score(y_test, y_predict))
print('f1_score' , f1_score(y_test, y_predict, pos_label='positive', average='weighted'))
print('precision_score', precision_score(y_test, y_predict, pos_label='positive', average='macro'))
print('recall_score' , recall_score(y_test, y_predict, pos_label='positive', average='micro'))
```
