```python
import pandas as pd

data = pd.read_csv ('emotion.csv')  
df = pd.DataFrame(data)
df
```

```python
x=pd.Series(df.Arabic_Term.values,index=df.Sentiment_Score).to_dict()
x
```

```python
positive_emt=[]
negitive_emt=[]
neutral_emt=[]
for key in x:
    if key > 0.5:
        positive_emt.append(x[key])
    elif key <-0.5:
        negitive_emt.append(x[key])
    else:
        neutral_emt.append(x[key])  
print(len(positive_emt))
positive_emt
```

```python
print(len(negitive_emt))
negitive_emt
```

```python
print(len(neutral_emt))
neutral_emt
```

```python
def count_neg(text):
    neg = 0
    for w in word_tokenize(text):
        if w in negitive_emt:
            neg +=1

    return neg
```

```python
def count_pos(text):
    pos = 0
    for w in word_tokenize(text):
        if w in positive_emt:
            pos +=1
            
    return pos
```

```python
def count_neutral(text):
    neu = 0
    for w in word_tokenize(text):
        if w in neutral_emt:
            neu +=1
            
    return neu
```

<h1>Original DataSet<h1>

```python
import pandas as pd

data = pd.read_csv ('DataSet.csv')   
df = pd.DataFrame(data)
print (df)
```

```python
def pre_process(df):
    df['pos'] = df['Tweet'].apply(count_pos)
    df['neg'] = df['Tweet'].apply(count_neg)
    df['neu'] = df['Tweet'].apply(count_neutral)
    return df
pre_process(df)
```

```python
X = df[['pos', 'neg','neu']]
```

```python
from sklearn.preprocessing import LabelEncoder
l = LabelEncoder()
y = l.fit_transform(df[['feeling']])
y
```

```python
def encode(text):
    if text == 'positive':
        return 2
    elif text =='negative':
        return 0
    else:
        return 1


y = df['feeling'].apply(encode)
y
```
